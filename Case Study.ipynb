```python
from tkinter import*
sc = Tk()
sc.title("Dữ liệu nhân viên")

sv_ten = StringVar()
sv_tuoi = StringVar()
sv_gioi_tinh = StringVar()
sv_que_quan = StringVar()


def nutnhan():
    f = open("nhanvien.csv","a",encoding="utf8")
    f.write(sv_ten.get() + "," + sv_tuoi.get() + "," + sv_gioi_tinh.get() + "," +  sv_que_quan.get() + "\n")
    f.close()
    sv_ten.set("")
    sv_tuoi.set("")
    sv_gioi_tinh.set("")
    sv_que_quan.set("")

def doichucnang():
    en_tuoi.config(state="disabled")
    en_gioi_tinh.config(state="disabled")
    en_que_quan.config(state="disabled")
    bt_luu.grid_forget()
    bt_tim.grid(row=6, column=1, columnspan = 2, pady = (20,0))
    bt_doi_chuc_nang.config(text = "Nhập", command = doichucnang1)
    
def doichucnang1():
    en_tuoi.config(state="normal")
    en_gioi_tinh.config(state="normal")
    en_que_quan.config(state="normal")
    bt_luu.grid(row=6, column=1, columnspan = 2, pady = (20,0))
    bt_tim.grid_forget()
    bt_doi_chuc_nang.config(text = "Tra Cứu", command = doichucnang)


def tim():
    f = open("nhanvien.csv","r",encoding="utf8")
    data = f.readlines()
    f.close()
 
    L_ten = []
    L_tuoi = []
    L_gioi_tinh = []
    L_que_quan = []
    
    for i in range(len(data)):
        data[i] = data[i].strip()
        data[i] = data[i].split(",")
        L_ten.append(data[i][0])
        L_tuoi.append(data[i][1])
        L_gioi_tinh.append(data[i][2])
        L_que_quan.append(data[i][3])
        
    if sv_ten.get() in L_ten:
        i = L_ten.index(sv_ten.get())
        sv_tuoi.set(L_tuoi[i])
        sv_gioi_tinh.set(L_gioi_tinh[i])
        sv_que_quan.set(L_que_quan[i])
        
            
lb = Label(sc,text = "Phần mềm nhập thông tin nhân viên", font = ("aria",15))
lb.grid(row=1,column=1,columnspan=2,pady = (0,20))
    
lb_ten = Label(sc, text="Tên", font = ("Arial", 12), fg = "red")
lb_ten.grid(row=2, column=1)

en_ten = Entry(sc, textvariable = sv_ten)
en_ten.grid(row=2, column=2)

lb_tuoi = Label(sc, text="Tuổi", font = ("Arial", 12), fg = "red")
lb_tuoi.grid(row=3, column=1)

en_tuoi = Entry(sc, textvariable = sv_tuoi)
en_tuoi.grid(row=3, column=2)

lb_gioi_tinh = Label(sc, text="Giới Tính", font = ("Arial", 12), fg = "red")
lb_gioi_tinh.grid(row=4, column=1)

en_gioi_tinh = Entry(sc, textvariable = sv_gioi_tinh)
en_gioi_tinh.grid(row=4, column=2)

lb_que_quan = Label(sc, text="Quê Quán", font = ("Arial", 12), fg = "red")
lb_que_quan.grid(row=5, column=1)

en_que_quan = Entry(sc, textvariable = sv_que_quan)
en_que_quan.grid(row=5, column=2)

bt_luu = Button(sc, text="Lưu", fg = "Red", font = ("Arial", 14), width = 10, command = nutnhan)
bt_luu.grid(row=6, column=1, columnspan = 2, pady = (20,0))

bt_tim = Button(sc, text="Tìm", fg = "Red", font = ("Arial", 14), width = 10, command = tim)

bt_doi_chuc_nang = Button(sc, text="Tra Cứu", fg = "Red", font = ("Arial", 14), width = 10, command = doichucnang)
bt_doi_chuc_nang.grid(row=7, column=1, columnspan = 2, pady = (0,30))



sc.mainloop()

```
